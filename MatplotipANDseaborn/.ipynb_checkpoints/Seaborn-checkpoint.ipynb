```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("winequality-red.csv")
print("___________SON 10 SATIR:___________\n",df.tail(5).T)
```

```python
df.info()#dataframe hakkında genel bilgiler sunar
print("\nTekrarlanan Satır Sayısı: ",df.duplicated().sum())# dataframe'deki yinelemeli  satırların sayısını verir.
#tekrarlanan sayıları kaldırıyoruz
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.describe().T
```

```python
sns.boxplot(x= df['quality'],color='#ff7d00')
plt.xlabel('Şarap Kalitesi')
plt.ylabel('Frequancy')
plt.title("Kalite sınıfı")
plt.show()
```

```python
plt.figure(figsize=(26,12))
sns.barplot(x='free sulfur dioxide', y='total sulfur dioxide', data=df,color='#ffdd00')
plt.xlabel('free sulfur dioxide')
plt.ylabel('total sulfur dioxide')
plt.title("Sülfür Dioksit ")
plt.show()
```

```python
# Dağılım Grafiği
plt.figure(figsize=(12,6))
palette = ["#03273c", "#ff99c8", "#c6f8e5", "#8400ff", "#013c58", "#cd001a"] #paleti özelleştirelim
sns.scatterplot(x='alcohol', y='pH', data=df, hue='quality', palette=palette,marker='>')
plt.title("Dağılım Grafiği")
plt.grid(True)
```

```python
# Distplot
sns.histplot(df['alcohol'], kde=True, color='#d0f4de')
plt.title("Distplot")
plt.show()

# KDEplot
sns.kdeplot(df['pH'], color='#936fac', fill=True)# fill ifadesi içinin boş veya boyalı olmasını sağlar
plt.title("KDEplot")
plt.show()
```

```python
# Korelasyon matrisi ile heatmap oluşturma
plt.figure(figsize=(10,10))
corr = df.corr()
sns.heatmap(corr ,annot=True ,linewidths=0.5, cmap= 'BuGn');
```
