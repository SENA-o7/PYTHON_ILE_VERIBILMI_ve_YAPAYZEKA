```python
import pandas as pd
data = pd.read_csv("iris.csv")
print("___________İLK 10 SATIR:___________\n",data.head(10))# iris verisinin ilk 10 veri setini görüntüledik
```

```python
#eksik veri var mı?
print("\nEksik veri var mı?:")
print(data.isnull().values.any())
```

```python
#Veri setini biraz anlamaya çalışalım .Veri seti özeti:
print("___________VERİ SETİ ÖZETİ:___________\n",data.describe().T)
```

```python
import matplotlib.pyplot as plt


#çizgi ve çubuk grafiği
plt.figure(figsize=(12,6)) #grafik tablosunun uzunluğu ,genişliği
plt.plot(data.index,data["sepal_length"],label='Sepal Length',color='purple',marker='o')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.title('_____Sepal Uzunluğu____')
plt.legend() #label  etikedini grafiğimizde gösteriri yani açıklama yapmamıza yarar
plt.grid(True)#arka plana çizgiler ekler
# grafiği kaydetme:
plt.savefig('irislength.png')
plt.show()    

plt.figure(figsize=(12,6)) #grafik tablosunun uzunluğu ,genişliği
plt.bar(data.index, data["sepal_width"] , label='Sepal Width',color='orange', alpha=0.7) #alpha =çubuğun şeffaflık düzeyi(0-1)
plt.xlabel('Index')
plt.ylabel('Genişlik (cm)')
plt.title('_______Sepal Genişliği______')
plt.legend()#label  etikedini grafiğimizde gösteriri yani açıklama yapmamıza yarar
plt.grid(True)#arka plana çizgiler ekler   
plt.savefig('iriswidth.png')
plt.show() 
```

```python
#alt grafik
fig, axs = plt.subplots(2, 1, figsize=(8, 10))#axs alt grafik eksenleri 2 satır 1 sütun içerin alt grafik:

axs[0].scatter(data["petal_length"], data["petal_width"], color='green', label='Petal Length vs Width',marker='<')
axs[0].set_title('Petal Uzunluğu ve Genişliği')
axs[0].set_xlabel('Petal Length (cm)')
axs[0].set_ylabel('Petal Width (cm)')
axs[0].legend()
axs[0].grid(True)

axs[1].hist(data["species"], bins=10, color='pink', alpha=0.7, label='Sepal Length Distribution')
axs[1].set_title('Sepal Uzunluğu Dağılımı')
axs[1].set_xlabel('Özellikler')
axs[1].set_ylabel('Frekans')
axs[1].legend()
axs[1].grid(False)

```
