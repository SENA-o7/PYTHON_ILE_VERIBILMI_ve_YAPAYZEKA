```python
import numpy as np
import matplotlib.pyplot as plt


dizi=np.linspace(0,2*np.pi,100)
sin_dizi=np.sin(dizi)
cos_dizi=np.cos(dizi)

#matplotlib ile grafikleri izdirme
plt.figure(figsize=(12,5))
plt.plot(dizi,sin_dizi,label="sin(x)",color="lightgreen")
plt.plot(dizi,cos_dizi,label="cos(x)",color="yellow")
plt.title("Sins ve Kosins Grafii")
plt.xlabel("dizi (radyan)")
plt.ylabel("Deer")
plt.axhline(0, color='orange', linewidth=2, linestyle="--")
plt.gca().spines['top'].set_color('darkgreen')
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_color('darkgreen')
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['bottom'].set_color('darkgreen')
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_color('darkgreen')
plt.gca().spines['left'].set_linewidth(2)
plt.legend()
plt.grid(alpha=0.6)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


dizi= np.linspace(0, 2 * np.pi, 100)
sin_dizi = np.sin(dizi)
cos_dizi = np.cos(dizi)

# Veriyi bir DataFrame'e koyma
data = pd.DataFrame({"x": dizi, "Sin(x)": sin_dizi, "Cos(x)": cos_dizi})

# Seaborn izimi
sns.lineplot(data=data, x="x", y="Sin(x)", label="Sin(x)", color="pink")
sns.lineplot(data=data, x="x", y="Cos(x)", label="Cos(x)", color="purple")

# Grafik detaylar
plt.title("Sins ve Kosins Grafii (Seaborn)")
plt.xlabel("dizi(radyan)")
plt.ylabel("Deer")

plt.axhline(0, color='darkblue', linewidth=1, linestyle="--")
plt.gca().spines['top'].set_color('blue')
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_color('blue')
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['bottom'].set_color('blue')
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_color('blue')
plt.gca().spines['left'].set_linewidth(2)
plt.grid(alpha=0.6)
plt.legend()
plt.show()
```
